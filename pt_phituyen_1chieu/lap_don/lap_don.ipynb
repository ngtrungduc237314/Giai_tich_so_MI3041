# PHUONG PHAP LAP DON 

# LUU Y: 

Doi voi phuong phap lap don, ta phai xac dinh cong thuc lap truoc roi moi kiem tra tinh hoi tu bang dieu kien hoi tu. 

Cu the $f(x) = 0 \Rightarrow x = \phi(x)$

Neu $\phi(x)$ khong thoa man dieu kien hoi tu thi phai chon lai $\phi(x)$ khac.

# A. Co so ly thuyet
## 1. Dieu kien 
1.1. $(a,b)$ la khoang cach ly nghiem 

1.2. $|\phi'(x)| \leq q < 1 \quad \forall x \in (a,b)$

(Co gang chon $q$ cang nho thi phuong phap hoi tu cang nhanh)

(Neu $\phi'(x) > 0 \forall x \in [a,b]$ thi day $x_k$ tien den nghiem dung tu 1 phia; nguoc lai, neu $\phi'(x) < 0 \forall x \in [a,b]$ thi day $x_k$ tien den nghiem dung theo 2 phia)

## 2. Cong thuc lap 
$x_{k} = \phi(x_{k-1})$ voi $k = 1,2,\dots$

## 3. Cong thuc sai so 
3.1. Cong thuc sai so tien nghiem: 
$ |x_k - \alpha| \leq \frac{q^k}{1 - q}|x_1 - x_0|$ voi $0 < q < 1$

3.2. Cong thuc sai so hau nghiem: 
$ |x_k - \alpha| \leq \frac{q}{1 - q}|x_k - x_{k-1}| $ voi $0 < q < 1$

## 4. Danh gia sai so & Dieu kien dung
4.1. Cho truoc so lan lap $\N_0$ $\Rightarrow$ danh gia sai so

(1) $ |x_k - \alpha| \leq \frac{q^k}{1 - q}|x_1 - x_0|$ voi $0 < q < 1$

(2) $ |x_k - \alpha| \leq \frac{q}{1 - q}|x_k - x_{k-1}| $ voi $0 < q < 1$

4.2. Cho truoc sai so tuyet doi $\varepsilon$ $\Rightarrow$ dieu kien dung

a. Cong thuc tien nghiem
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{q^k}{1 - q}|x_1 - x_0| \leq \varepsilon
&\Rightarrow \frac{q^k}{1 - q}|x_1 - x_0| \leq \varepsilon\\
&\Rightarrow k \geq \log_{q} {\frac{\varepsilon(1-q)}{|x_1 - x_0|}} 
\end{aligned}
$$ 
Chon k = $[\log_{q} {\frac{\varepsilon(1-q)}{|x_1 - x_0|}}] + 1.$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{q}{1 - q}|x_k - x_{k-1}| < \varepsilon
&\Rightarrow \frac{q}{1 - q}|x_k - x_{k-1}| < \varepsilon \\
&\Rightarrow |x_k - x_{k-1}| < \varepsilon \frac{q}{1 - q}
\end{aligned}
$$

4.3. Cho truoc sai so tuong doi $\varepsilon$ $\Rightarrow$ dieu kien dung: cong thuc hau nghiem
$$
\begin{aligned}
\frac{|x_k - \alpha|}{|x_k|} \leq \frac{q}{1 - q}\frac{|x_k - x_{k-1}|}{|x_k|} < \varepsilon 
&\Rightarrow \frac{q}{1 - q}\frac{|x_k - x_{k-1}|}{|x_k|} < \varepsilon \\
&\Rightarrow \frac{|x_k - x_{k-1}|}{|x_k|} < \varepsilon \frac{q}{1 - q}
\end{aligned}
$$

# B. Thuat toan & danh gia sai so 
## 1. Cho truoc sai so tuyet doi $\varepsilon$
### 1.1. Dieu kien dung
a. Cong thuc tien nghiem
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{q^k}{1 - q}|x_1 - x_0| \leq \varepsilon
&\Rightarrow \frac{q^k}{1 - q}|x_1 - x_0| \leq \varepsilon\\
&\Rightarrow k \geq \log_{q} {\frac{\varepsilon(1-q)}{|x_1 - x_0|}} 
\end{aligned}
$$ 
Chon k = $[\log_{q} {\frac{\varepsilon(1-q)}{|x_1 - x_0|}}] + 1.$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{q}{1 - q}|x_k - x_{k-1}| < \varepsilon
&\Rightarrow \frac{q}{1 - q}|x_k - x_{k-1}| < \varepsilon \\
&\Rightarrow |x_k - x_{k-1}| < \varepsilon \frac{q}{1 - q}
\end{aligned}
$$

### 1.2. Thuat toan 

(Giao trinh phuong phap tinh - tr.42)

*INPUT*: $\phi$ - ham lap; $x_0$ - xap xi ban dau; $a,b$ - cac diem bien cua khoang cach ly nghiem; $\varepsilon$ - sai so tuyet doi; $N_0$ - so lan lap toi da.

*OUTPUT*: $x$ - nghiem gan dung cua phuong trinh $f(x) = 0$. 

*Thuat giai*:

**Bước 1**: 
Kiem tra $(a,b)$ la khoang cach ly nghiem cua $f(x)$ hay khong. Neu khong thoa man, dung chuong trinh.

**Bước 2**: 
Tinh $q$ = $\max\limits_{x \in [a,b]} |\phi'(x)|$.

**Bước 3**: 
Neu $q > 1$ thi dung chuong trinh.

**Bước 4**:
Gan $i = 1$ va $ \varepsilon = \frac{\varepsilon(1 - q)}{q}$.

**Bước 5**:
Neu $i \leq N_0$ thi thuc hien **Buoc 5.1 - 5.3**
- Buoc 5.1. 
Tinh $x = \phi(x_0)$.
- Buoc 5.2.
Neu $|x - x_0| < \varepsilon$ thi in ra $x$. Dung chuong trinh. 
- Buoc 5.3.
Dat $i = i + 1$ va $x_0 = x$. Quay lai **Buoc 5**

**Bước 6:** In ra x. Dung chuong trinh do so lan lap dat toi da. 

### 1.3. Danh gia sai so 
Sai so tuyet doi $\Delta x = \frac{q}{1 - q}|x - x_0| + 0.5 \times 10^{-rbl} \leq \varepsilon$ voi $rbl$ la so chu so thap phan cua nghiem gan dung $x$.
