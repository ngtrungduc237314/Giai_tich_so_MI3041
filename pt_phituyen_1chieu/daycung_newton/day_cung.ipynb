# PHUONG PHAP DAY CUNG 

# A. Co so ly thuyet
## 1. Dieu kien 
1.1. $(a,b)$ la khoang cach ly nghiem 

1.2. $f', f''$ xac dinh dau tren $(a,b)$ 

1.3. Chọn điểm đầu $x_0$, điểm đích $d$ sao cho $f(x_0)f''(x_0) < 0$, $d$ là điểm còn lại.  

(1.1, 1.2 la dieu kien thuc hien phuong phap; ca 3 la dieu kien hoi tu cua phuong phap) 

## 2. Cong thuc lap 
$x_{k} = x_{k-1} - \frac{f(x_{k-1})(d - x_{k-1})}{f(d) - f(x_{k-1})}$ voi $k = 1,2,\dots$

## 3. Cong thuc sai so 
3.1. Cong thuc sai so muc tieu: 

$|x_k - \alpha| \leq \frac{|f(x_k)|}{m_1}$ voi $m_1 = \min_{x \in [a,b]} |f'(x)|$.

3.2. Cong thuc sai so hau nghiem .II: 

$|x_k - \alpha| \leq \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}|$ 

voi $m_1 = \min_{x \in [a,b]} |f(x)|$, $M_1 = \max_{x \in [a,b]} |f(x)|$. 

## 4. Danh gia sai so 
4.1. Cho truoc so lan lap $N_0$ $\Rightarrow$ danh gia sai so

(1) $|x_k - \alpha| \leq \frac{|f(x_k)|}{m_1}$ voi $m_1 = \min_{x \in [a,b]} |f'(x)|$

(2) $|x_k - \alpha| \leq \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}|$

4.2. Cho truoc sai so tuyet doi $\varepsilon$ $\Rightarrow$ dieu kien dung

a. Cong thuc muc tieu
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{|f(x_k)|}{m_1} \leq \varepsilon
&\Rightarrow \frac{|f(x_k)|}{m_1} \leq \varepsilon \\
\end{aligned}
$$ 
voi $m_1 = \min_{x \in [a,b]} |f'(x)|$.

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| < \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}| < \varepsilon
&\Rightarrow \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}| < \varepsilon \\
&\Rightarrow |x_{k} - x_{k-1}| < \varepsilon \frac{m_1}{M_1 - m_1} 
\end{aligned}
$$
voi $m_1 = \min_{x \in [a,b]} |f'(x)|$, $M_1 = \max_{x \in [a,b]} |f'(x)|$. 

4.3. Cho truoc sai so tuong doi $\varepsilon$ $\Rightarrow$ dieu kien dung

a. Cong thuc muc tieu
$$
\begin{aligned}
\frac{|x_k - \alpha|}{|x_k|} \leq \frac{|f(x_k)|}{m_1 \times |x_k|} < \varepsilon
\end{aligned}
$$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
\frac{|x_k - \alpha|}{|x_k|} \leq \frac{M_1 - m_1}{m_1 \times |x_k|}|x_{k} - x_{k-1}| < \varepsilon
\end{aligned}
$$

# B. Thuat toan & danh gia sai so 
## 1. Cho truoc sai so tuyet doi $\varepsilon$
### 1.1. Dieu kien dung
a. Cong thuc muc tieu
$$
\begin{aligned}
|x_k - \alpha| \leq \frac{|f(x_k)|}{m_1} \leq \varepsilon
&\Rightarrow \frac{|f(x_k)|}{m_1} \leq \varepsilon \\
\end{aligned}
$$ 
voi $m_1 = \min_{x \in [a,b]} |f'(x)|$.

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| < \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}| < \varepsilon
&\Rightarrow \frac{M_1 - m_1}{m_1}|x_{k} - x_{k-1}| < \varepsilon \\
&\Rightarrow |x_{k} - x_{k-1}| < \varepsilon \frac{m_1}{M_1 - m_1} 
\end{aligned}
$$
voi $m_1 = \min_{x \in [a,b]} |f'(x)|$, $M_1 = \max_{x \in [a,b]} |f'(x)|$. 

### 1.2. Thuat toan 

(Giao trinh phuong phap tinh - tr.32)

*INPUT*: $f(x)$ - ham so; $a,b$ - cac diem bien cua khoang cach ly nghiem; $\varepsilon$ - sai so; $N_0$ - so lan lap toi da.

*OUTPUT*: $x$ - nghiem gan dung cua phuong trinh $f(x) = 0$. 

*Thuat giai*:

**Bước 1**: 
Kiem tra 2 dieu kien dau bao gom [a,b] la khoang cach ly nghiem va $f', f''$ xac dinh dau tren [a,b]. 

**Bước 2**: 
Neu $f(a)f''(a) < 0$, dat $x_0 = a, d = b$. Trai lai, dat $x_0 = b, d = a$.

**Bước 3**: 
Tinh $m_1 = \min_{x \in [a,b]} {|f'(x)|}$ va $M_1 = \max_{x \in [a,b]} {|f'(x)|}$.
Gan $\varepsilon := \frac{m_1\varepsilon}{M_1 - m_1}$ va gan $i = 1$.

**Bước 4**:
Neu $i \leq N_0$ thi thuc hien **Buoc 4.1 - 4.3**
- Buoc 4.1. 
Tinh $dx = -{\frac{f(x_0)(x_0 - d)}{f(x_0 - d)}}$ va $x_1 = x_0 + dx$.
- Buoc 4.2.
Neu $|dx| < \varepsilon$ thi $x = x_1$. Dung chuong trinh. 
- Buoc 4.3.
Dat $i = i + 1$ va $x_0 = x_1$. Quay lai **Buoc 4.1.**

**Bước 5:** Nghiem cua phuong trinh $x = x_1$.
Dung chuong trinh do so lan lap dat toi da. 

### 1.3. Danh gia sai so 
Sai so tuyet doi $\Delta x = \frac{M_1 - m_1}{m_1}|x_1 - x_0| + 0.5 \times 10^{-rbl} \leq \varepsilon$ voi $rbl$ la so chu so thap phan cua nghiem gan dung $x$.

## 3. Cho truoc sai so tuong doi $\delta$
### 3.1. Dieu kien dung: Cong thuc hau nghiem 
$$
\begin{aligned}
\frac{|x_k - \alpha|}{|x_k|} \leq \frac{|a_{k} - b_{k}|}{|x_k|} < \delta 
&\Rightarrow \frac{|a_{k} - b_{k}|}{|x_k|} < \delta
\end{aligned}
$$
