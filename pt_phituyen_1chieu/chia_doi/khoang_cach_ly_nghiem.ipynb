# Khoang cach ly nghiem


# 1. DN: 
$(a,b)$ duoc goi la mot khoang cach ly nghiem neu $f(x)$ chi co duy nhat 1 nghiem $\alpha$ trong khoang $(a,b)$ do.

# 2. Cach tim kcln $(a,b)$: 
a. (Phuong phap giai tich) Dieu kien du: Neu $f(x)$ la mot ham so lien tuc, don dieu tren $(a,b)$ va $f(a)f(b) < 0$ thi $(a,b)$ la khoang cach ly nghiem cua ham so $f(x)$. 

b. (Phuong phap hinh hoc) Ve do thi & xac dinh cac kcln.
