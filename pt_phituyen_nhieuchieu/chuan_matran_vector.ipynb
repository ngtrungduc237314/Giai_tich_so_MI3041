# Chuan cua ma tran va vector
## 1. Chuan cua vector (chinh la chuan cua ma tran m x 1)
Cho vector X xac dinh boi 
$$
X =
\begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{pmatrix}
$$

1.1. Chuan hang (chuan vo cung)

$$
\|X\|_\infty = \max_{1 \le i \le n} \{|x_i|\} = \max \{|x_1|, |x_2|, \dots, |x_n|\}
$$

1.2. Chuan cot (chuan 1)

$$
\|X\|_1 = \sum_{i=1}^{n} |x_i| = |x_1| + |x_2| + \ldots + |x_n|$$

## 2. Chuan cua ma tran 
Cho ma tran $A$ xac dinh boi 
$$
A = (a_{ij})_{m \times n}
$$

2.1. Chuan hang (chuan vo cung)
$$ \|A\|_\infty = \max_{1 \le i \le m} \sum_{j=1}^{n} |a_{ij}| = \max_{1 \le i \le m} \{|a_{i1}| + |a_{i2}| + \dots + |a_{in}|\}$$

2.2. Chuan cot (chuan 1)
$$
\|A\|_1 = \max_{1 \le j \le n} \sum_{i=1}^{m} |a_{ij}| = \max_{1 \le j \le n} \{|a_{1j}| + |a_{2j} + \dots + |a_{mj}|\}
$$
