# PHUONG PHAP LAP DON 

# Bai toan

Cho
$$
F(X) = 0 \Leftrightarrow 
\begin{cases}
f_1(x_1, x_2, \ldots, x_n) = 0, \\
f_2(x_1, x_2, \ldots, x_n) = 0, \\
\vdots \\
f_n(x_1, x_2, \ldots, x_n) = 0.
\end{cases}
$$

Tim mot nghiem gan dung $X$ cua he phuong trinh phi tuyen nay.

# Y tuong: 
$$
X = \Phi(X) \Leftrightarrow
\begin{cases}
x_1 = \varphi_1(x_1, x_2, \ldots, x_n), \\
x_2 = \varphi_2(x_1, x_2, \ldots, x_n), \\
\vdots \\
x_n = \varphi_n(x_1, x_2, \ldots, x_n).
\end{cases}
$$


# A. Co so ly thuyet
## 1. Dieu kien 
1.1. Chon mien cach ly nghiem $D = [a_1, b_1] \times [a_2, b_2] \times \dots \times [a_n, b_n]$ thoa man: 

a) $\Phi(X)$ lien tuc tren mien $D$ 

b) $\Phi(X) \in D \quad \forall X \in D$.

1.2. Chon $K$ thoa man 1 trong 3 truong hop sau: 

a) $K = \max_{1 \le i \le n, 1 \le j \le n} \left| \frac{\partial \varphi_i}{\partial x_j} \right| \times n < 1$ 

b) $K = \max_{1 \le i \le n} \sum_j \left| \frac{\partial \varphi_i}{\partial x_j} \right| < 1 (p = \infty)$

c) $K = \max_{1 \le j \le n} \sum_i \left| \frac{\partial \varphi_i}{\partial x_j} \right| < 1 (p = 1)$


(Co gang chon $\min q$  de phuong phap hoi tu nhanh nhat)

## 2. Cong thuc lap 
$$
X = \Phi(X) \Leftrightarrow
\begin{cases}
x_1 = \varphi_1(x_1, x_2, \ldots, x_n), \\
x_2 = \varphi_2(x_1, x_2, \ldots, x_n), \\
\vdots \\
x_n = \varphi_n(x_1, x_2, \ldots, x_n).
\end{cases}
$$

## 3. Cong thuc sai so 
3.1. Cong thuc sai so tien nghiem: 
$ \|X_k - \alpha\|_p \leq \frac{K^k}{1 - K}\|X_1 - X_0\|_p$ voi $0 < K < 1$

3.2. Cong thuc sai so hau nghiem: 
$ \|X_k - \alpha\|_p \leq \frac{K}{1 - K}\|X_k - X_{k-1}\|_p $ voi $0 < K < 1$

## 4. Danh gia sai so & Dieu kien dung
4.1. Cho truoc so lan lap $\N_0$ $\Rightarrow$ danh gia sai so

(1) $ \|X_k - \alpha\|_p \leq \frac{K^k}{1 - K}\|X_1 - X_0\|_p$ voi $0 < K < 1$

(2) $ \|X_k - \alpha\|_p \leq \frac{K}{1 - K}\|X_k - X_{k-1}\|_p $ voi $0 < K < 1$

4.2. Cho truoc sai so tuyet doi $\varepsilon$ $\Rightarrow$ dieu kien dung

a. Cong thuc tien nghiem
$$
\begin{aligned}
\|X_k - \alpha\|_p \leq \frac{K^k}{1 - K}\|X_1 - X_0\|_p \leq \varepsilon
&\Rightarrow \frac{K^k}{1 - K}\|X_1 - X_0\|_p \leq \varepsilon\\
&\Rightarrow k \geq \log_{K} {\frac{\varepsilon(1-K)}{\|X_1 - X_0\|_p}} 
\end{aligned}
$$ 
Chon k = $[\log_{K} {\frac{\varepsilon(1-K)}{\|X_1 - X_0\|_p}}] + 1.$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
\|X_k - \alpha\|_p \leq \frac{K}{1 - K}\|X_k - X_{k-1}\|_p < \varepsilon
&\Rightarrow \frac{K}{1 - K}\|X_k - X_{k-1}\|_p < \varepsilon \\
&\Rightarrow \|X_k - X_{k-1}\|_p < \varepsilon \frac{K}{1 - K}
\end{aligned}
$$

4.3. Cho truoc sai so tuong doi $\varepsilon$ $\Rightarrow$ dieu kien dung: cong thuc hau nghiem
$$
\begin{aligned}
\frac{\|X_k - \alpha\|_p}{\|X_k\|_p} \leq \frac{K}{1 - K}\frac{\|X_k - X_{k-1}\|_p}{\|X_k\|_p} < \varepsilon 
&\Rightarrow \frac{K}{1 - K}\frac{\|X_k - X_{k-1}\|_p}{\|X_k\|_p} < \varepsilon \\
&\Rightarrow \frac{\|X_k - X_{k-1}\|_p}{\|X_k\|_p} < \varepsilon \frac{K}{1 - K}
\end{aligned}
$$

# B. Thuat toan & danh gia sai so 
## 1. Cho truoc sai so tuyet doi $\varepsilon$
### 1.1. Dieu kien dung
a. Cong thuc tien nghiem
$$
\begin{aligned}
\|X_k - \alpha\|_p \leq \frac{K^k}{1 - K}\|X_1 - X_0\|_p \leq \varepsilon
&\Rightarrow \frac{K^k}{1 - K}\|X_1 - X_0\|_p \leq \varepsilon\\
&\Rightarrow k \geq \log_{K} {\frac{\varepsilon(1-K)}{\|X_1 - X_0\|_p}} 
\end{aligned}
$$ 
Chon k = $[\log_{K} {\frac{\varepsilon(1-K)}{\|X_1 - X_0\|_p}}] + 1.$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
\|X_k - \alpha\|_p \leq \frac{K}{1 - K}\|X_k - X_{k-1}\|_p < \varepsilon
&\Rightarrow \frac{K}{1 - K}\|X_k - X_{k-1}\|_p < \varepsilon \\
&\Rightarrow \|X_k - X_{k-1}\|_p < \varepsilon \frac{K}{1 - K}
\end{aligned}
$$

### 1.2. Thuat toan 

*INPUT*: $n$ - so phan tu $x_i$ can tim, $\varphi_1, \varphi_2, \dots, \varphi_n$ - ham lap cho tung toa do cua ham lap $\Phi(X)$; $x_1, x_2, \dots, x_n$ - cac toa do cua xap xi ban dau X_0; $a_1, b_1, a_2, b_2, \dots, a_n, b_n$ - cac diem bien cua mien cach ly nghiem; $\varepsilon$ - sai so tuyet doi; $N_0$ - so lan lap toi da.

*OUTPUT*: $X$ - nghiem gan dung cua phuong trinh $F(X) = 0$. 

*Thuat giai*:

**Bước 1**: 
Kiem tra $D = [a_1, b_1] \times [a_2, b_2] \times \dots [a_n, b_n]$ la mien cach ly nghiem cua $F(X)$ hay khong. Neu khong thoa man, dung chuong trinh.

**Bước 2**: 
- **Buoc 2.1.1.** Tinh $K_2 = \max_{1 \le i \le n} \sum_j \left| \frac{\partial \varphi_i}{\partial x_j} \right| < 1$ va gan $p := \infty$.

- **Buoc 2.1.2.** Tinh $K_3 = \max_{1 \le j \le n} \sum_i \left| \frac{\partial \varphi_i}{\partial x_j} \right| < 1$ va gan $p := 1$.

- **Buoc 2.1.3.** Neu $K_2 < K_3$ thi gan $K = K_2$ va $p = \infty$. Nguoc lai, gan $K = K_3$ va $p = 1$. 

**Bước 3**: 
Neu $K > 1$ thi dung chuong trinh.

**Bước 4**:
Gan $i = 1, p = \infty$.  va $ \varepsilon = \frac{\varepsilon(1 - K)}{K}$.

**Bước 5**:
Neu $i \leq N_0$ thi thuc hien **Buoc 5.1 - 5.3**
- **Buoc 5.1** 
$X = F(X_0)$
- **Buoc 5.2**
Neu $\|X - X_0\|_p < \varepsilon$ thi in ra $X$. Dung chuong trinh. 
- **Buoc 5.3.**
Dat $i = i + 1$ va $X_0 = X$. Quay lai **Buoc 5**

**Bước 6:** In ra $X$. Dung chuong trinh do so lan lap dat toi da. 

### 1.3. Danh gia sai so 
Sai so tuyet doi $\Delta x = \frac{K}{1 - K}\|X - X_0\|_p \leq \varepsilon$.

# C. Vi du minh hoa
$$
\begin{cases}
3x_1 - \cos x_2x_3 - 0.5 = 0 \\
x_1^2 - 81(x_2 + 0.1)^2 + \sin x_3 + 1.06 = 0 \\
e^{-x_1x_2} + 20x_3 + 9.1389 = 0
\end{cases}
$$

$$
\Leftrightarrow
\begin{cases}
x_1 = \dfrac{1}{3} (\cos x_2x_3 + 0.5) \\
x_2 = \dfrac{1}{9} \sqrt{x_1^2 + \sin x_3 + 1.06} - 0.1 \\
x_3 = -\dfrac{1}{20} e^{-x_1x_2} - \dfrac{9.1389}{20}
\end{cases}
$$

Tim nghiem $X = (x_1, x_2, x_3)$.

## Giai:

**1. Với hàm $\varphi_1 = \dfrac{1}{3}(\cos(x_2x_3) + 0.5)$:**

* $\left|\dfrac{\partial \varphi_1}{\partial x_1}\right| = 0$
* $\left|\dfrac{\partial \varphi_1}{\partial x_2}\right| = \left|-\dfrac{1}{3}x_3 \sin(x_2x_3)\right| \le \dfrac{1}{3} \cdot 1 \cdot \sin(1) \approx 0.2805$
* $\left|\dfrac{\partial \varphi_1}{\partial x_3}\right| = \left|-\dfrac{1}{3}x_2 \sin(x_2x_3)\right| \le \dfrac{1}{3} \cdot 1 \cdot \sin(1) \approx 0.2805$

**2. Với hàm $\varphi_2 = \dfrac{1}{9}\sqrt{x_1^2 + \sin x_3 + 1.06} - 0.1$:**

* $\left|\dfrac{\partial \varphi_2}{\partial x_1}\right| = \left|\dfrac{x_1}{9\sqrt{x_1^2 + \sin x_3 + 1.06}}\right| \le \dfrac{1}{9\sqrt{1^2 + \sin(-1) + 1.06}} \approx 0.1006$
* $\left|\dfrac{\partial \varphi_2}{\partial x_2}\right| = 0$
* $\left|\dfrac{\partial \varphi_2}{\partial x_3}\right| = \left|\dfrac{\cos x_3}{18\sqrt{x_1^2 + \sin x_3 + 1.06}}\right| \le \dfrac{\cos(-1)}{18\sqrt{0^2 + \sin(-1) + 1.06}} \approx 0.0640$

**3. Với hàm $\varphi_3 = -\dfrac{1}{20}e^{-x_1x_2} - \dfrac{9.1389}{20}$:**

* $\left|\dfrac{\partial \varphi_3}{\partial x_1}\right| = \left|\dfrac{x_2}{20}e^{-x_1x_2}\right| \le \dfrac{1}{20}e^1 \approx 0.1359$
* $\left|\dfrac{\partial \varphi_3}{\partial x_2}\right| = \left|\dfrac{x_1}{20}e^{-x_1x_2}\right| \le \dfrac{1}{20}e^1 \approx 0.1359$
* $\left|\dfrac{\partial \varphi_3}{\partial x_3}\right| = 0$

Ma tran Jacobi duoc xac dinh nhu sau:

$$
J = \begin{bmatrix} 
0 & 0.2805 & 0.2805 \\ 
0.1006 & 0 & 0.0640 \\ 
0.1359 & 0.1359 & 0 
\end{bmatrix}
$$

#### Trường hợp 1: Ước lượng lỏng (Công thức $n \times \max$)
$$K = \max_{i,j} \left| \dfrac{\partial \varphi_i}{\partial x_j} \right| \times 3$$

* **Ý nghĩa:** Đây là cách ước lượng đơn giản nhất. Ta lấy phần tử lớn nhất trong toàn bộ ma trận $M$, sau đó nhân với bậc của hệ (ở đây hệ có 3 phương trình nên $n=3$).
* **Thực tế tính toán:** Theo ma trận $M$ đã lập, phần tử lớn nhất là $0.2805$. 
  $$\Rightarrow K = 0.2805 \times 3 = 0.8415$$
*(Lưu ý: Trên slide ghi $0.238 \times 3$, có thể do tác giả khảo sát trên miền hẹp hơn hoặc sai số làm tròn. Cách làm chuẩn vẫn là lấy phần tử max nhân với $n$)*.

#### Trường hợp 2: Chuẩn hàng / Chuẩn vô cùng ($p = \infty$)
$$K = \max_i \sum_j \left| \dfrac{\partial \varphi_i}{\partial x_j} \right|$$

* **Ý nghĩa:** Tính tổng trị tuyệt đối của các phần tử theo từng hàng ngang, sau đó chọn ra giá trị lớn nhất.
* **Cách tính cụ thể:**
    * Tổng hàng 1 ($i=1$): $0 + 0.2805 + 0.2805 = \mathbf{0.5610}$
    * Tổng hàng 2 ($i=2$): $0.1006 + 0 + 0.0640 = 0.1646$
    * Tổng hàng 3 ($i=3$): $0.1359 + 0.1359 + 0 = 0.2718$
* **Kết quả:** $\max(0.5610, 0.1646, 0.2718) = 0.561$. Vì $K = 0.561 < 1$, hệ đảm bảo hội tụ.

#### Trường hợp 3: Chuẩn cột ($p = 1$)
$$K = \max_j \sum_i \left| \dfrac{\partial \varphi_i}{\partial x_j} \right|$$

* **Ý nghĩa:** Tính tổng trị tuyệt đối của các phần tử theo từng cột dọc, sau đó chọn ra giá trị lớn nhất.
* **Cách tính cụ thể:**
    * Tổng cột 1 ($j=1$): $0 + 0.1006 + 0.1359 = 0.2365$
    * Tổng cột 2 ($j=2$): $0.2805 + 0 + 0.1359 = \mathbf{0.4164}$
    * Tổng cột 3 ($j=3$): $0.2805 + 0.0640 + 0 = 0.3445$
* **Kết quả:** $\max(0.2365, 0.4164, 0.3445) = 0.4164$. Làm tròn ta được **$0.416$** như trong slide.
