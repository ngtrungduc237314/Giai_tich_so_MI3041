# PHUONG PHAP CHIA DOI 

# A. Co so ly thuyet
## 1. Dieu kien 
1.1. $(a,b)$ la khoang cach ly nghiem <br>
1.2. $f(x)$ lien tuc tren $(a,b)$ <br>
1.3. $f(a)f(b) < 0$ <br>
## 2. Cong thuc lap 
$x_{k} = \frac{a_{k} + b_{k}}{2}$ voi $k = 0,1,2,\dots$
## 3. Cong thuc sai so 
3.1. Cong thuc sai so tien nghiem: 
$ |x_k - \alpha| \leq \frac{b-a}{2^k}$

3.2. Cong thuc sai so hau nghiem: 
$ |x_k - \alpha| \leq |a_{k} - b_{k}| $

## 4. Danh gia sai so & Dieu kien dung
4.1. Cho truoc so lan lap $\N_0$ $\Rightarrow$ danh gia sai so

(1) $ |x_k - \alpha| \leq \frac{b-a}{2^k}$ 

(2) $ |x_k - \alpha| \leq |a_{k} - b_{k}| $.

4.2. Cho truoc sai so tuyet doi $\varepsilon$ $\Rightarrow$ dieu kien dung

a. Cong thuc tien nghiem
$$
\begin{aligned}
|(x_k - \alpha)| \leq \frac{b-a}{2^k} \leq \varepsilon
&\Rightarrow \frac{b-a}{2^k} \leq \varepsilon\\
&\Rightarrow k \geq \log_{2} {\frac{b-a}{\varepsilon}}
\end{aligned}
$$ 
Chon k = $[\log_{2} {\frac{b-a}{\varepsilon}}] + 1$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| \leq |a_{k} - b_{k}| < \varepsilon
&\Rightarrow |a_{k} - b_{k}| < \varepsilon
\end{aligned}
$$

4.3. Cho truoc sai so tuong doi $\varepsilon$ $\Rightarrow$ dieu kien dung: cong thuc hau nghiem
$$
\begin{aligned}
\frac{|x_k - \alpha|}{|x_k|} \leq \frac{|a_{k} - b_{k}|}{|x_k|} < \varepsilon
&\Rightarrow \frac{|a_{k} - b_{k}|}{|x_k|} < \varepsilon \\
&\Rightarrow \frac{|a_{k-1} - b_{k-1}|}{|a_{k-1} + b_{k-1}|} < \varepsilon
\end{aligned}
$$

# B. Thuat toan & danh gia sai so 
## 1. Cho truoc so lan lap $\N_0$
### 1.1. Dieu kien dung: du $N_0$ iteration

### 1.2. Thuat toan ($x_n = (a_n + b_n)/2$)
**INPUT:** $a,b$ – các điểm biên của khoảng cách ly nghiệm; $\varepsilon > 0$ - sai so; $N_0$ – số lần lặp tối đa, $f(x)$ – hàm số.

**OUTPUT:** $c$ – nghiệm gần đúng của phương trình.

*Bước 1*:
Nếu $f(a)f(b) > 0$ thì dừng chương trình và thông báo  
“Điều kiện thực hiện phương pháp chia đôi không thỏa mãn”.

Nếu $f(a) = 0$ thì $\alpha = a$. Dừng chương trình.  

Nếu $f(b) = 0$ thì $\alpha = b$. Dừng chương trình.

*Bước 2*: 
Đặt $i = 1$, $f_a = f(a)$.

*Bước 3*: 
Trong khi $i \le N_0$ thực hiện Bước 3.1 – 3.4.

- Bước 3.1. 
Tính  
$$
c = \frac{a+b}{2}, \qquad f_c = f(c)
$$

- Bước 3.2.
Nếu $f_c = 0$ thì $\alpha = c$. Dừng chương trình.

- Bước 3.3. 
Nếu $f_a f_c > 0$ thì đặt $a = c$; ngược lại đặt $b = c$.

- Bước 3.4.
Gán $i = i + 1$.

*Bước 4*:
$\alpha = c$. Dừng chương trình do số lần lặp đạt tối đa.

### 1.3. Danh gia sai so 
Sai so tuyet doi $\Delta x = \frac{b-a}{2^{N_0}} + 0.5 \times 10^{-rbl} \leq \varepsilon$ voi $rbl$ la so luong chu so thap phan hay thu nguyen nho nhat cua nghiem gan dung $\alpha$.

## 2. Cho truoc sai so tuyet doi $\varepsilon$
### 2.1. Dieu kien dung 
a. Cong thuc tien nghiem
$$
\begin{aligned}
|(x_k - \alpha)| \leq \frac{b-a}{2^k} \leq \varepsilon
&\Rightarrow \frac{b-a}{2^k} \leq \varepsilon\\
&\Rightarrow k \geq \log_{2} {\frac{b-a}{\varepsilon}}
\end{aligned}
$$ 
Chon k = $[\log_{2} {\frac{b-a}{\varepsilon}}] + 1$

b. Cong thuc hau nghiem 
$$
\begin{aligned}
|x_k - \alpha| \leq |a_{k} - b_{k}| < \varepsilon
&\Rightarrow |a_{k} - b_{k}| < \varepsilon
\end{aligned}
$$
