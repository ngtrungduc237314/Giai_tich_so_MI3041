# Thuật toán: Phương pháp Gauss-Jordan

# Thuat toan va danh gia sai so

## Đầu vào
- Số nguyên dương $n$ (số phương trình và số ẩn)
- Ma trận mở rộng $A \in \mathbb{R}^{n \times (n+1)}$, với phần tử $a_{i,j}$

## Đầu ra
- Nghiệm $x_1, x_2, \dots, x_n$  
- Hoặc thông báo: *Hệ không có nghiệm duy nhất*

---

## Bước 1: Vòng lặp chính
Cho $i = 1 \rightarrow n$, thực hiện các bước sau:

---

## Bước 2: Tìm phần tử trụ
Tìm $p$ nhỏ nhất sao cho:
$$
i \le p \le n \quad \text{và} \quad a_{p,i} \ne 0
$$

- Nếu **không tồn tại** $p$:
  
  Thong bao:  *Hệ không có nghiệm duy nhất*  
  Dừng thuật toán

---

## Bước 3: Hoán vị hàng
Nếu $p \ne i$, đổi chỗ hàng $p$ và hàng $i$:

$$
\text{For } k = 1 \rightarrow n+1:
$$
$$
\begin{cases}
\text{temp} = a_{p,k} \\
a_{p,k} = a_{i,k} \\
a_{i,k} = \text{temp}
\end{cases}
$$

---

## Bước 4: Khử Gauss-Jordan
Với mọi $j = 1 \rightarrow n$, $j \ne i$:

- Tính hệ số:
$$
m = \frac{a_{j,i}}{a_{i,i}}
$$

- Cập nhật hàng $j$:
$$
\text{For } k = 1 \rightarrow n+1:
$$
$$
a_{j,k} = a_{j,k} - m \cdot a_{i,k}
$$

---

## Bước 5: Trích xuất nghiệm
Sau khi ma trận trở thành dạng đường chéo:

$$
x_i = \frac{a_{i,n+1}}{a_{i,i}}, \quad i = 1 \rightarrow n
$$

---

## Bước 6: Kết thúc
- In nghiệm:
$$
(x_1, x_2, \dots, x_n)
$$
- Kết thúc thuật toán
